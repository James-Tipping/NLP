# An Analysis of tweets about flight experience

## Classification of tweets as positive, negative, or neutral

## Using SVCs, naive Bayes', and logistic regression models

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('../DATA/airline_tweets.csv')
```

## Data preprocessing

```python
data = df[['airline_sentiment', 'text']]
data
```

```python
X = data['text']
y = data['airline_sentiment']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf.fit(X_train)
X_train_tfidf = tfidf.transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
X_train_tfidf
```

## Naive Bayes training

```python
nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)
```

## Logistic regression training

```python
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train_tfidf, y_train)
```

# Support vector classifiers training

```python
rbf_svc = SVC()
rbf_svc.fit(X_train_tfidf, y_train)
```

```python
linear_svc = LinearSVC()
linear_svc.fit(X_train_tfidf, y_train)
```

## Results of each model

```python
def report(model):
    preds = model.predict(X_test_tfidf)
    print(classification_report(y_test, preds))
    ConfusionMatrixDisplay.from_predictions(y_test, preds)
```

```python
report(linear_svc)
```

```python
report(log_model)
```

```python
report(rbf_svc)
```

```python
report(nb)
```

## Train on all data for model deployment

```python
pipe = Pipeline(steps=[('tfidf', TfidfVectorizer()), ('svc', LinearSVC())])
pipe.fit(X, y)
```
